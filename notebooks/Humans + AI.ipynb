## Interpretability vs explainability

**Interpretability** - "_What_ is in this model, _how_ does it treat input"


**Explainability** - "_Why_ did this model do that this time"

Here are some principles for providing good interpretation and explanation ...

**Interpretation/Explanation methods should be** 
* _Portable_ to other models
* _Expressive_ in presentation
* _Translucent_ in relying on model internals
* _Low in Algorithmic Complexity_

**Individual interpretation / explanations should have** 
* _Accuracy_ similar to model output (correctness)
* _Fidelity_ in approximation of model (resolution)
* _Consistency_ between models 
* _Stability_ between prediction instances
...

(...continued)
* _Comprehensibility_ for humans
* _Certainty_ for impact assessment
* _Importance_ for parts of explanation
* _Measure of Novelty_ of prediction instance
* _Representativeness_ in instance coverage

These are generally properties of related algorithms, but they also apply to UX of AI

## Example based explanations

### Counterfactual Explanations

A causal explanation:

`If X had not occurred, Y would not have occurred`


`If I hadn’t taken a sip of this hot coffee, I wouldn’t have burned my tongue`


`If your 2-hour serum insulin level was 154.3, you would have a score of 0.51`

`If you don't eat yer meat, you can't have any pudding 
How can you have any pudding if you don't eat yer meat?`

Pros
* Clear
* Doesn't require access to model / data, just prediction
* Works with non-ML systems
* Basically a loss function, fairly simple to implement

Cons
* Rashomon effect: you can have multiple matching explanations
* No guarantee an explanation instance can be found
* Lacks a general software solution

Growing Spheres algorithm
1. Make a sphere around your prediction in the latent space
2. Grab a sample outside, see if it's different
3. Make it bigger until the stuff you're grabbing is different

```python
from aif360.datasets import GermanDataset
from german_credit_data_reason_codes import reason_map, display_description
import german_credit_data_reason_codes
from sklearn.model_selection import train_test_split
```

```python
## load data
data = GermanDataset()
X, _ = data.convert_to_dataframe()
X.drop(['credit'], axis=1, inplace=True)
y = data.scores.ravel()
```

```python
## Split the data for model training / testing
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.33, 
                                                    random_state=42)
```

```python
## Investigate data
X.describe()
# https://pair-code.github.io/facets/index.html#facets-overview
```

```python
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(X_train, y_train)
y_predicted = model.predict(X_test)
y_probability = model.predict_proba(X_test)
```

```python
%matplotlib inline
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

fpr, tpr, thresholds = roc_curve(y_test, y_probability[:,1])
roc_auc = auc(fpr, tpr)

fig, ax = plt.subplots()
ax.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % (roc_auc))
ax.plot([0, 1], [0, 1], 'k--')
ax.set_xlim([0.0, 1.0])
ax.set_ylim([0.0, 1.05])
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.set_title('Receiver operating characteristic')
ax.legend(loc="lower right")
 
# create the axis of thresholds (scores)
ax2 = ax.twinx()
ax2.plot(fpr, thresholds, markeredgecolor='r',linestyle='dashed', color='r')
ax2.set_ylabel('Threshold',color='r')
ax2.set_ylim([thresholds[-1],thresholds[0]])
ax2.set_xlim([fpr[0],fpr[-1]]);
```

```python
import shap
shap.initjs()

# explain the model's predictions using SHAP values
# (same syntax works for LightGBM, CatBoost, and scikit-learn models)
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_test)
```

```python
"""
A14 : no checking account (Status of existing checking account)
A34 : critical account/ other credits existing (not at this bank) (Credit history)
A11 : ... < 0 DM (Status of existing checking account)
"""

# visualize the first prediction's explanation
shap.force_plot(explainer.expected_value, shap_values[0,:], X_test.iloc[0,:])
```

## Privacy

In Homogeneity Attacks<sup> [1](https://en.wikipedia.org/wiki/L-diversity)</sup> sensitive values can be extracted by leveraging similarity

In Background Knowledge Attacks <sup>[2](https://dl.acm.org/citation.cfm?doid=1217299.1217302)</sup> associations are used reduce the set of possible values for the sensitive attribute 

Models may contain PII and if so may need to be retrained to remove such data.

## Designing with AI in mind

Donald Norman, author of "The Design of Everyday Things" suggests designing to these principles:

* **visibility** - what you see
* **affordance** - cues for interaction
* **mapping** - connecting control and effect
* **feedback** - response to interaction
* **constraints** - what you can't do
* **consistancy** - helping users learn from patterns

These still apply to ML.

## Human-Centered Machine Learning

#### Don’t expect Machine learning to figure out what problems to solve

"if you aren’t aligned with a human need, you’re just going to build a very powerful system to address a very small — or perhaps nonexistent — problem."

<center><img src="https://thumbs.gfycat.com/BowedRipeComet-small.gif"></center>

#### Ask yourself if ML will address the problem in a unique way

The task below can be solved with `contains keyword`
<center><img src="https://cdn-images-1.medium.com/max/1000/1*URC8DMjzMCrWrzpjIddBVg.png" width=80%></center>

Describe the way a theoretical human “expert” might perform the task today.

If your human expert were to perform this task, how would you respond to them so they improved for the next time? Do this for all four phases of the confusion matrix.

If a human were to perform this task, what assumptions would the user want them to make?

Conduct [Wizard of Oz experiments](https://en.wikipedia.org/wiki/Wizard_of_Oz_experiment) where the AI is replaced with humans until its purpose is concrete.

#### Weigh the costs of false positives and false negatives

<center>How users may view a confusion matrix.

The 4 corners of the 'confusion matrix' will have different presentations and impacts on your users, but can usually be traded off.

If you cannot tolerate a missed crime, maybe you can tolerate a misfired warning.

If a misfired warning is bad, by using explainability and usability testing you might mitigate the negative impact.

#### Plan for 'Human in the loop' and adaptation

<center>Your users may form inaccurate understandings of your models.</center>

<center><img src="anthropomorphic.jpg"></center>

Labels can't be collected until you know what you need. 

And you don't know what you need until you've trained a model.

Allow for 'truthiness' in labeling and plan to spend some time iterating on your labels
